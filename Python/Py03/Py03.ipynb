```python
import math

# --- 연산자 선택 ---
def get_operator_input():
    """유효한 연산자 입력 후, 사용자에게 재확인을 받아 최종 확정하는 함수"""
    # 연산자 기호와 한글 이름을 매핑한 딕셔너리
    op_names = {
        '+': '더하기',
        '-': '빼기',
        '*': '곱하기',
        '/': '나누기',
        '**': '제곱'
    }

    while True:
        # 1. 연산자 입력 받기
        op = input(f"연산자를 입력하세요 ({', '.join(op_names.keys())}): ").strip()

        # 2. 유효한 연산자인지 1차 검증
        if op in op_names:
            # 3. 사용자에게 연산자 명을 호출하며 재확인
            confirm = input(f"입력하신 연산자가 '{op_names[op]}({op})'가 맞습니까? (y/n): ").lower().strip()

            if confirm == 'y':
                print(f"👉 '{op_names[op]}' 연산으로 확정되었습니다.")
                return op  # 최종 확정되어 함수 종료
            else:
                print("연산자 선택을 취소했습니다. 다시 입력해주세요.")
        else:
            print(f"오류: '{op}'는 지원하지 않는 연산자입니다. 다시 입력해주세요.")

# --- 메인 계산기 실행 ---
def run_calculator():
    print("=== 파이썬 간단 계산기 프로그램을 시작합니다 ===")

    while True:
        # 1. 숫자 입력 (기존 함수 활용)
        num1 = get_integer_input("\n첫 번째 정수를 입력하세요: ")
        num2 = get_integer_input("두 번째 정수를 입력하세요: ")

        # 2. 연산자 입력 및 확정 (수정된 부분)
        operator = get_operator_input()

        # 3. 계산 로직 (연산자가 이미 확정되었으므로 else 처리가 필요 없음)
        try:
            if operator == '+':
                result = num1 + num2
            elif operator == '-':
                result = num1 - num2
            elif operator == '*':
                result = num1 * num2
            elif operator == '/':
                if num2 == 0:
                    raise ZeroDivisionError("0으로 나눌 수 없습니다.")
                result = num1 / num2
            elif operator == '**':
                result = math.pow(num1, num2)

            print(f"👉 계산 결과: {num1} {operator} {num2} = {result}")

        except ZeroDivisionError as e:
            print(f"오류: {e}")

        # 4. 종료 여부 확인
        cont = input("\n계속 계산하시겠습니까? (종료: n / 계속: 아무키): ").lower()
        if cont == 'n':
            print("계산기를 종료합니다. 이용해주셔서 감사합니다!")
            break #계산기 종료

if __name__ == "__main__":
    run_calculator()
```
