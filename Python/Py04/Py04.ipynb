# 사각형 넓이 구하기

```python
a = Square()

class Square:
    def __init__(self):
        self.square = int(input('넓이를 구하고 싶은 사각형의 숫자를 써주세요.\n 1.직사각형 2.평행사변형 3.사다리꼴 \n >>>'))

        if self.square == 1:
            print('직사각형 함수는 rect()입니다.')

        elif self.square == 2:
            print('평행사변형 함수는 par()입니다.')
        
        elif self.square == 3:
            print('사다리꼴 함수는 trape()입니다.')
        
        else:
            print('1, 2, 3 중에서 다시 입력해주세요')

    def rect(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '직사각형의 넓이는 : ' + str(area)
        return result

    def par(self):
        width, vertical = map(int, input('가로, 세로를 입력하세요. 예시 : 가로,세로\n >>>').split(','))
        area = width * vertical
        result = '평행사변형의 넓이는 : ' + str(area)
        return result

    def trape(self):
        width1, width2, vertical = map(int, input('윗변, 아랫변, 세로를 입력하세요. 예시 : 윗변, 아랫변,세로\n >>>').split(','))
        area = ((width1 + width2) * vertical) / 2
        result = '사다리꼴의 넓이는 : ' + str(area)
        return result
```

# Kiosk

```python
menu = ['americano', 'latte', 'mocha', 'yuza_tea', 'green_tea', 'choco_latte']
price = [2000, 3000, 3000, 2500, 2500, 3000]

class Kiosk:
    def __init__(self):
        self.menu = menu
        self.price = price
        
        # 주문 내역을 저장할 리스트 초기화
        self.order_menu = []
        self.order_price = []
        self.price_sum = 0
        self.temp = ""

    def menu_print(self):
        for i in range(len(self.menu)):
            print(i + 1, self.menu[i], ' : ', self.price[i], '원')

    # 주문 메서드 (추가 주문 로직 포함)
    def menu_select(self):
        print() 
        n = -1  # 초기값 설정 (while문 진입을 위해)
        
        # 추가 주문 또는 지불 루프
        while n != 0:
            print()
            n = int(input("추가 주문은 음료 번호를, 지불은 0을 누르세요 : "))
            
            # 메뉴판에 있는 번호를 입력했을 때
            if 0 < n <= len(self.menu):
                # 음료 온도 선택 루프
                t = 0
                while t != 1 and t != 2:
                    t = int(input("HOT 음료는 1을, ICE 음료는 2를 입력하세요 : "))
                    if t == 1:
                        self.temp = "HOT"
                    elif t == 2:
                        self.temp = "ICE"
                    else:
                        print("1과 2 중 하나를 입력하세요.\n")
                
                # 데이터 업데이트 및 리스트 추가 (영수증 출력을 위해 필수!)
                self.order_price.append(self.price[n-1])
                self.order_menu.append(self.temp + ' ' + self.menu[n-1])
                self.price_sum += self.price[n-1]
                
                # 현재 주문 상태 출력
                print(f'추가 주문 음료: {self.temp} {self.menu[n-1]} : {self.price[n-1]}원')
                print(f'현재 합계 : {self.price_sum}원')
            
            else:
                if n == 0:
                    print("주문이 완료되었습니다.")
                else:
                    print("없는 메뉴입니다. 다시 주문해 주세요.")

    def pay(self):
        print(f'\n총 합계 금액은 {self.price_sum}원 입니다.')
        self.pay_method = 0
        while self.pay_method != 1 and self.pay_method != 2:
            self.pay_method = int(input('현금 결제는 1번, 카드 결제는 2번을 입력해주세요. : '))
            if self.pay_method == 1:
                print('직원을 호출하였습니다.')
            elif self.pay_method == 2:
                print('IC칩 방향에 맞게 카드를 꽂아주세요.')
            else:
                print('다시 결제를 시도해 주세요.')

    def table(self):
        outline1 = "-" + "+" + "-" * 30 + "+" + "-"
        outline2 = "|" + " " * 31 + "|"
        
        print(outline1)
        for _ in range(3):
            print(outline2)
            
        # 리스트에 쌓인 모든 주문 내역 출력
        for i in range(len(self.order_menu)):
            print(f"  {self.order_menu[i]} : {self.order_price[i]}원")
            
        print(f'  합계 금액 : {self.price_sum}원')
        
        for _ in range(3):
            print(outline2)
        print(outline1)
```

```python
a = Kiosk()  # 객체 생성 
a.menu_print()  # 메뉴 출력
a.menu_select()  # 주문
a.pay()  # 지불
a.table()  # 주문표 출력
```
